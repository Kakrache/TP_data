```python
#supprimer les valeurs nulls, les vides , les doublons
df = df.dropna()
df = df.drop_duplicates()
df = df[df["Price"]>0]   #vérifier sil y ades pris annormale
df["Date"] = pd.to_datetime(df["Date"], errors="coerce") #convertir les date en format date
df['Date'] = df['Date'].dt.date #extraire la date

df
```

```python
#convertir les colonnes en bon format
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce', downcast='integer')
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df.dtypes
#enregistrer une copie du datset nettoyer
data = df.to_csv("data_clean.csv", index = False)
```

```python
#calculer le chiffre d'affaire total
df.head()
df['total'] = df["Price"]* df["Quantity"]
chiffre_affaire_total = df["total"].sum()
print(f"Le chiffre d'affaires total est : {chiffre_affaire_total}")
```

```python
#calcule du produit le plus vendu avec la quantité vendu
prod_vendu = df.groupby("Product")["Quantity"].sum()  
prod_plus_vendu = prod_vendu.idxmax()
Quantitu_max = prod_vendu.max()
print(f"Le produit le plus vendu est '{prod_plus_vendu}' avec {Quantitu_max} unités vendues.")
```

```python
#crée le resumer des ventes des prosuits 
resume_vente = df.groupby("Product").agg(
    total_quantity = ("Quantity", 'sum'),
    total_CA = ("total", "sum")
).reset_index()
with open('resume_ventes_par_produit.csv', mode='w', newline='', encoding='utf-8') as file:
    resume_vente.to_csv(file, index=False)
    
```
